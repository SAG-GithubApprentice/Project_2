### Use this Jupyter Script to dreate a Dataframe with current 10-year historical data for Intel, AMD and NVIDIA.

```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta

# Define the stock tickers
tickers = ['INTC', 'AMD', 'NVDA']

# Calculate the date ten years ago from today
end_date = datetime.now()
start_date = end_date - timedelta(days=10*365)

# Create an empty list to store the dataframes
dfs = []

# Fetch historical data for each stock and append to the list
for ticker in tickers:
    data = yf.download(ticker, start=start_date, end=end_date)
    data['Ticker'] = ticker  # Add a column for the ticker symbol
    dfs.append(data)

# Concatenate the dataframes into a single dataframe
all_data = pd.concat(dfs)

# Save the DataFrame to a CSV file
all_data.to_csv('stock_data_10_years.csv')

# Print the first few rows of the DataFrame
print(all_data.head())
```
